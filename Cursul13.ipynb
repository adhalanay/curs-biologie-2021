## Analiza datelor numerice

Cel mai adesea rezultatele experimentelor vor fi măsurători sau serii de măsurători. Este foarte important să putem extrage informații din aceste rezultate numerice. Vom vedea cum putem să prelucrăm o cantitate mare de numere pentru a obține concluzii despre obiectele pe care le studiem. 

Să presupunem că vrem să vedem cum influențează cantitatea de îngrășămînt creșterea unei plante. Pentru aceasta putem semăna într-un ghiveci al cărui pămînt a fost fertilizat un bob de fasole și în alt ghiveci fără îngrășămînt alt bob de fasole (controlul). După 20 de zile măsurăm plantele care au răsărit și găsim că prima a crescut la 15 cm, iar a doua 10 cm. Noi știm însă că creșterea plantelor este influențată de mai mulți factori deci nu putem trage concluzii doar din aceste două exemplare. 

Să presupunem că facem același experiment cu cîte cinci boabe în fiecare grup și obținem

Control    | 9,3 14,2 11,7 10,2 9,8 | 
-----------|------------------------|
Experiment |12,1 16,3 13,2 13,5 14,9|

Se vede că avem o variabilitate a lungimilor, una dintre plantele de control a fost mai înaltă decît 3 din cele fertilizate. 

Datele variază din două motive principale: în primul rînd sînt erorile experimentale (cantitate de îngrășămînt nu este cea corectă, rigla este mai puțin precisă sau maăsurătoarea nu este făcută corect etc.) apoi creșterea unei plante depinde de un număr mare de factori și boabele de fasole pot să difere din punct de vedere genetic, de asemenea condițiile de sol, lumină și aer pot fi diferite în ciuda eforturilor noastre.

Deci în studiul fenomenelor biologice este de așteptat să întîlnim o variabilitate a datelor obținute, chiar dacă ne străduim să minimizăm erorile experimentale. Ca să înțelegem variabilitatea rezultatelor avem nevoie să facem multe măsurători astfel încît să obținem cît mai multe date. 

Problema fundamentală a statisticii este că prea puține date ne pot conduce la concluzii greșite, dar prea multe devin greu de înțeles. De aceea ar trebui să grupăm datele ca să le facem mai ușor de înțeles.

Să presupunem că avem acum 20 de plante în grupul de control avînd lungimile 9,3, 9,7, 10,1, 10,4, 10,7, 10,9, 11, 11,1, 11,1, 11,3, 11,3, 11,6, 11,7, 11,9, 12,3, 13,4. Grupînd plantele în intervale de lungime 1 și obținem histograma de mai jos, care nu spune că sînt 2 plante în intervalul 9-10,
5 în intervalul 10-11, 8 în intervalul 11-12, 2 în intervalul 12-13, 1 în intervalul 13-14.

```julia
using Plots
pyplot()
x=[9.3, 9.7, 10.1, 10.2, 10.4, 10.6, 10.7, 10.7, 10.9, 11.0, 11.1, 11.1, 11.3, 11.3, 11.6, 11.7, 11.9, 12.3, 12.4, 13.4]
histogram(x,label="Lungimi")
```

```julia
using Plots
pyplot()
x=[9.3, 9.7, 10.1, 10.2, 10.4, 10.6, 10.7, 10.7, 10.9, 11.0, 11.1, 11.1, 11.3, 11.3, 11.6, 11.7, 11.9, 12.3, 12.4, 13.4]
histogram(x,bins=10,label="Lungimi")
```

A dou figură se obține luînd intervale de lungime 0,5. Numărul de puncte în fiecare interval este dat de aria barei corespunătoare. Dacă reducem din ce în ce mai mult grosimea barelor în histogramă, vîrfurile se aliniază pe graficul

```julia
using Plots
pyplot()
f(t)=exp(-t^2/2)
plot(f,-3,3)
```

Funcția al cărei grafic l-am desenat se numește distribuția datelor. În cazul nostru se numește **distribuția normală**. Evident că dacă facem din ce în ce mai multe măsurători și vom lua intervale din ce mai fine, atunci vîrfurile histogramelor se vor potrivi din ce în ce mai bine cu graficul funcției.

### Media, Mediana și Modulul

Un prim pas pentru prelucrarea datelor este să desenăm histograma. Apoi vrem să determinăm tendința centrală a datelor: datele se acumulează în jurul unei valori centrale sau nu. Dacă apare că se strînge în jurul unei valori care este aceea? Avem trei valori care ne ajută să studiem aceste întrebări.

**Modulul** este valoarea care apare cel mai frecvent. Această definiție este mai greu de folosit, deoarece s-ar putea ca mai multe valori să apară cu aceeași frecvență (ca în exemplul nostru). De aceea este mai convenabil să spunem că pentru un interval de lungime 1, modulul se află în intervalul $[11,12)$ sau pentru un interval de lungime 0,5 între $[11,11,5).$ 

**Mediana** este valoarea aflată fix în poziția din mijloc atunci cînd toate valorile sînt puse în ordine crescătoare. În exemplul nostru, avem un număr par de numere, iar la mijloc se află 11 și 11,1. Facem media lor și obținem că mediana este 11,05.
